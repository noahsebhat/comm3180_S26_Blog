I'm not getting married any time soon (at least I don't think I am), but I sometimes think about what my ideal wedding would be like. An important component of that vision is the wedding date. There's a lot of factors that dictate a good wedding date, like venue and guest availability, but perhaps the most important is weather. Using publicly available Philadelphia weather data, I looked to determine which single date would be the best for me to get married in Philadelphia, given my weather and wedding type preferences.

I've always thought an outdoor wedding would be cool, so it's important that it does not rain on the day of my wedding. Thus, my first step in determining my ideal wedding day was finding which has been the dryest day in the last 20 years.

Set up

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Loading the data

```python
weather_df = pd.read_csv('data/philadelphia_weather_2005_to_2025.csv')
weather_df = weather_df.assign(date=pd.to_datetime(weather_df['date']))
```

Subsetting days in the data that have had no recorded rain

```python
no_rain = weather_df[weather_df["rain"] == 0]
```

Identifying commonly ocurring day in the subset

```python
dryest_day = no_rain["day"].value_counts().idxmax()
print(dryest_day)
```

Counting the amount of times it has not rained on that day

```python
dryest_day_df = no_rain[no_rain["day"] == dryest_day]
count = len(dryest_day_df)
print(count)
```

Since 2005, August 26th has been the dryest day, not recording a single day of rain. That's great! But as a San Francisco native, I've grown up around moderate weather and don't like it when it's too hot or cold. So next, I wanted to find what the average temperature on August 26th has been over the last 20 years.

Caluclating average temperature on August 26th

```python
avg_temp = (dryest_day_df["high"] + dryest_day_df["low"]).mean() / 2
avg_temp = round(avg_temp, 1)
print(avg_temp)
```

The average temperature on August 26th has been 77 degrees farenheight. That's literally perfect. However, out of caution, I was interested in further examining the temperature data to make sure that the average temperature on August 26th hasn't been trending too far up or down.

Plotting the data

```python
dryest_day_df["avg_temp"] = (dryest_day_df["high"] + dryest_day_df["low"]) / 2
plt.scatter(dryest_day_df["year"], dryest_day_df["avg_temp"])
plt.xlabel("Year")
plt.ylabel("Average Temperature")
plt.title("Average Temperature on August 26th (2005 - 2025)")
plt.show()
```

Looking at the data, there doesn't appear to a meaningful trend one way or the other so August 26th still checks out! 

Of course, no date comes with a weather guarantee, but if historical data is any guide, August 26th has some of the best odds for dry, warm weather in Philadelphia. So save the date! If and when I get married in Philadelphia, it will be on August 26th.
